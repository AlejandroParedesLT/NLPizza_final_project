```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
from sklearn.pipeline import make_pipeline
```

```python
df = pd.read_csv("/Users/jahnavimaddhuri/Downloads/2017_1.csv")
df.dropna(subset=["body", "political_leaning"], inplace=True)
df.head()
```

```python
# split the into train/validation
x = df["body"]  
y = df["political_leaning"]
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
pipeline = make_pipeline(
    TfidfVectorizer(max_features=50000, ngram_range=(1, 2), stop_words="english"),
    MultinomialNB(alpha=0.1)
    # alpha can be used to adjust smoothing
)

# train
pipeline.fit(x_train, y_train)

# validate
y_pred = pipeline.predict(x_val)

# evaluate
print("Classification Report:")
print(classification_report(y_val, y_pred, target_names=["LEFT", "RIGHT", "CENTER", "UNDEFINED"]))

print(f"Accuracy: {accuracy_score(y_val, y_pred):.2f}")
```
